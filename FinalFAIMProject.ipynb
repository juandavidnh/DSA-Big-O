```python
import pandas as pd

# List of your datasets (replace with your actual file paths or dataframes)
datasets = ['ListingsJan2024.csv', 'ListingsFeb2023.csv', 'ListingsMar2023.csv', 'ListingsApr2023.csv',
            'ListingsMay2023.csv', 'ListingsJun2023.csv', 'ListingsJul2023.csv', 'ListingsAug2023.csv',
            'ListingsSep2023.csv', 'ListingsOct2023.csv', 'ListingsNov2023.csv', 'ListingsDec2023.csv']

months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

# Read each dataset and add the month column
all_data = []
for i, dataset in enumerate(datasets):
    df = pd.read_csv(dataset)
    df['Month'] = months[i]
    all_data.append(df)

# Concatenate all dataframes into one
master_data = pd.concat(all_data, ignore_index=True)

# Now master_data is your combined dataset with a Month column
```
